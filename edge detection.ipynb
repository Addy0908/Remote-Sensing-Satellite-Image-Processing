```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("C://Users//adrij//OneDrive//Desktop//adrija//vtop//sem5//ip//DA//ComputerVision-Practice-main//spectralEnhancement_3//satellite.jpeg",0)
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#corner detection
import cv2
import numpy as np
 
filename = "C://Users//adrij//OneDrive//Desktop//adrija//vtop//sem5//ip//DA//ComputerVision-Practice-main//spectralEnhancement_3//satellite.jpeg"
img = cv2.imread(filename)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
 
gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)
 
#result is dilated for marking the corners, not important
dst = cv2.dilate(dst,None)
 
# Threshold for an optimal value, it may vary depending on the image.
img[dst>0.01*dst.max()]=[0,0,255]
 
cv2.imshow('haris_corner', img)
cv2.waitKey()
```

```python
#importing the required libraries
#Extracting Edge Features
import numpy as np

from skimage.io import imread, imshow
from skimage.filters import prewitt_h,prewitt_v
import matplotlib.pyplot as plt
%matplotlib inline

#reading the image 
image = imread("C://Users//adrij//OneDrive//Desktop//adrija//vtop//sem5//ip//DA//ComputerVision-Practice-main//spectralEnhancement_3//satellite.jpeg",as_gray=True)

#calculating horizontal edges using prewitt kernel
edges_prewitt_horizontal = prewitt_h(image)
#calculating vertical edges using prewitt kernel
edges_prewitt_vertical = prewitt_v(image)

imshow(edges_prewitt_vertical, cmap='gray')
```
